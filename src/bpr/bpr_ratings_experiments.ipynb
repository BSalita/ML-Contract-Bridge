```python
import pandas as pd
import pathlib
from collections import defaultdict
import mlBridgeLib
```

```python
rootPath = pathlib.Path('e:/bridge/data')
acblPath = rootPath.joinpath('acbl')
acblPath.mkdir(parents=True,exist_ok=True)
bprPath = rootPath.joinpath('bpr')
bprPath.mkdir(parents=True,exist_ok=True)
```

```python
# read all files downloaded from BridgePowerRatings and merge into a single df.
files = list(bprPath.glob('*.pkl'))
dfs = {}
names = []
for n,f in enumerate(files):
    df_read = pd.read_pickle(bprPath.joinpath(f))
    display(f'Processing {n+1}/{len(files)}: file:{f} len:{len(df_read)}',df_read.head())
    if 'Name' in df_read:
        assert 'player_id' in df_read
        dfs[f.stem] = df_read
        names += df_read['Name'].values.tolist()
unique_names = sorted(set(names))
f'names:{len(names)} unique_names:{len(unique_names)}'
```

```python
df = pd.DataFrame({'Name':unique_names})        
display(len(df),df)
```

```python
# merge bpr files on 'Name'. Column names are tuples in preparation for converting to MultiIndex
# todo:
# if 'Name' is assigned a wrong player_id, the data will be inconsistent e.g. differing Unit for same player.
df = pd.DataFrame({'Name':unique_names})        
for n,(k,df_read) in enumerate(dfs.items()):
    display(f'Processing {n+1}/{len(files)}: file:{k} len:{len(df_read)}',df_read.head())
    assert 'Name' in df_read
    assert 'player_id' in df_read
    if k.startswith('d'): # player should only be a member of one district so don't use district number
        for col in df_read.columns:
            if col in ['Name','player_id']:
                continue
            df[('district',col)] = pd.merge(df,df_read[['Name',col]],on='Name',how='left')[col]
            df = df.rename({'district_'+col:('district',col)}, level=0)
    else:
        for col in df_read.columns:
            if col in ['Name','player_id']:
                continue
            df[(k,col)] = pd.merge(df,df_read[['Name',col]],on='Name',how='left')[col]
    display('df:',len(df),df.head())
print(f'Done: Total files processed:{len(dfs)} len:{len(df)}')
```

```python
# convert df into MultiIndex
names = df.pop('Name')
#df = df[sorted(df.columns)]#,key=lambda c: c[0])]
df.columns = pd.MultiIndex.from_tuples(df.columns)
df.insert(0, ('','Name'), names)
display(df)
```

```python
df.to_excel(bprPath.joinpath('bpr_merged_tables.xlsx'))
```

```python
# takes 2s
# drop rows where ns_match_points + ew_match_points are outliers.
# todo: revisit
sum_rounded = (df['ns_match_points']+df['ew_match_points']).round(1) # round(1) or round(2)?
sum_rounded_dict = sum_rounded.value_counts().to_dict()
#df['ns_ew_match_point_vc'] = sum_rounded.map(sum_rounded_dict)
drop_match_points = sum_rounded.eq(0) | sum_rounded.map(sum_rounded_dict).le(200)
print(len(df),sum(drop_match_points))
df[drop_match_points]
```

```python
# takes 2m30s
Club = df['club_id_number'].astype(str)
mae = [p[1][0] for p in df['club_session'].str.split(' ')] # Morning, Afternoon, Evening
Event = df['game_date'].dt.strftime('%y%m%d').astype(str)+mae
HandRecord = df['hand_record_id']
Board = df['board_number'].astype(str).str.zfill(2)
Key = Club+'_'+Event+'_'+Board
ClubDate = Club+'_'+Event
EventBoard = Event+'_'+Board
NSPair = df['section_name'].astype(str)+df['ns_pair'].astype(str).str.zfill(2)
EWPair = df['section_name'].astype(str)+df['ew_pair'].astype(str).str.zfill(2)
Pair = NSPair+'_'+EWPair
NNum = [p[0] for p in df['player_number_ns']]
NName = [p[0] for p in df['player_name_ns']]
SNum = [p[1] for p in df['player_number_ns']]
SName = [p[1] for p in df['player_name_ns']]
ENum = [p[0] for p in df['player_number_ew']]
EName = [p[0] for p in df['player_name_ew']]
WNum = [p[1] for p in df['player_number_ew']]
WName = [p[1] for p in df['player_name_ew']]
PairNS = ['_'.join(p) for p in df['player_number_ns']]
PairEW = ['_'.join(p) for p in df['player_number_ew']]
MP_NS = df['mp_total_ns']
MP_EW = df['mp_total_ew']
Score = df['ns_score']
Par = df['board_id'].map(pard)
MatchP = df['ns_match_points']
Pct = df['ns_match_points']/(df['ns_match_points']+df['ew_match_points']).round(1) # round(1) or round(2)?
contract = df['contract'].str.replace(' ','').str.upper().str.replace('NT','N') # remove spaces. force upper-case. replace NT with N.
contractdf = contract.str.extract(r'^(?P<BidLvl>\d)(?P<BidSuit>C|D|H|S|N)(?P<Dbl>X*)$')
BidLvl = contractdf['BidLvl']
BidSuit = contractdf['BidSuit']
Dbl = contractdf['Dbl']
NSEW = df['declarer']
Tricks = df['result'] # df['tricks_taken'] # -1, = +1, ...
EventBoardScore = EventBoard+'_'+Score
```

```python
# attempting to free up some memory.
del df
import gc
gc.collect()
```

```python
# takes 10s
Declarer = []
OnLead = []
Dummy = []
NotOnLead = []
for i,(d,n,e,s,w) in enumerate(zip(NSEW,NNum,ENum,SNum,WNum)):
    assert i == len(Declarer) and i == len(OnLead) and i == len(Dummy) and i == len(NotOnLead)
    if d == 'N':
        Declarer.append(n)
        OnLead.append(e)
        Dummy.append(s)
        NotOnLead.append(w)
    elif d == 'E':
        Declarer.append(e)
        OnLead.append(s)
        Dummy.append(w)
        NotOnLead.append(n)
    elif d == 'S':
        Declarer.append(s)
        OnLead.append(w)
        Dummy.append(n)
        NotOnLead.append(e)
    elif d == 'W':
        Declarer.append(w)
        OnLead.append(n)
        Dummy.append(e)
        NotOnLead.append(s)
    else: # d == nan
        Declarer.append(None)
        OnLead.append(None)
        Dummy.append(None)
        NotOnLead.append(None)
        #print(i,d,n,e,s,w)
```

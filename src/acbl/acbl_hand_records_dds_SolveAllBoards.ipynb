```python
# obsolete?
# valdiate calculated dd result by comparing against known (assumed) correct dd. Similar to functions.CompareTable().
def CompareDDTables(DDtable1, DDtable2):
    for suit in range(dds.DDS_STRAINS):
        for pl in range(4):
            if DDtable1[suit][pl] != DDtable2[4 * suit + pl]:
                return False
    return True
```

```python
# obsolete?
# valdiate calculated par result by comparing against known (assumed) correct par. Similar to functions.ComparePar().
def ComparePars(par1, par2):
    if par1[0] != par2[0]:
        return False
    if len(par1[1]) != len(par2[1]):
        return False
    for p1,p2 in zip(sorted(par1[1],key=lambda k: k[1]),sorted(par2[1],key=lambda k: k[1])):
        if p1 != p2:
            if p1[1] != p2[1] or p1[2] != p2[2] or p1[3] != p2[3]:
                return False # suit/double/direction difference
            if p1[0]+p1[4] != p2[0]+p2[4]:
                return False # only needs to have trick count same. ignore if levels are same (min level+overs vs max level+0)
    return True
```

```python
# takes ?h per ?k.
# perform initialization for analysis of play of the hand.

import hands
import dds
import ctypes
import functions

starting_handno = 0 # for restarting at a particular hando (0-len(stdhrs)). normally 0.

bo = dds.boardsPBN()
solved = dds.solvedBoards()
chunkSize = 1

dds.SetMaxThreads(0)

nhands = 3 #len(stdhrs)
max_boards = dds.MAXNOOFBOARDS
```

```python
# Analyse play of the hand. Pass groups of max_boards to maximize throughput. Restartable.
# todo: implement direction in pbn. always using 'N:' for now.

for n,grp_start in enumerate(range(starting_handno,nhands,max_boards)):
    n += 1
    grp_count = min(nhands-grp_start,max_boards)
    bo.noOfBoards = grp_count
    print(f'Processing group:{n} hands:{grp_start} to {grp_start+grp_count} of {nhands}')
    DDboards = []
    for handno in range(grp_count):
        r = stdhrs.iloc[grp_start+handno]
        nsew_tuple = r['Hands']
        pbn = 'N:'+' '.join('.'.join([suit for suit in suits]) for suits in nsew_tuple)
        assert len(pbn) == 1+1+13*4+3*4+3, len(pbn) # 69=='N'+':'+(13 cards per hand)+(3 '.' per suit per hand)+3 spaces
        pbn_encoded = pbn.encode()
        #print(len(pbn),pbn,pbn_encoded)

        bo.deals[handno].trump = list(mlBridgeLib.SHDC).index('S')
        bo.deals[handno].first = list(mlBridgeLib.NESW).index('N')

        bo.deals[handno].currentTrickSuit[0] = 0
        bo.deals[handno].currentTrickSuit[1] = 0
        bo.deals[handno].currentTrickSuit[2] = 0
        
        bo.deals[handno].currentTrickRank[0] = 0
        bo.deals[handno].currentTrickRank[1] = 0
        bo.deals[handno].currentTrickRank[2] = 0

        bo.deals[handno].remainCards = pbn_encoded # original hand
        
        bo.deals[handno].target = -1
        bo.deals[handno].solutions = 3
        bo.deals[handno].mode = 0
        bo.target[handno] = -1
        bo.solutions[handno] = 3
        bo.mode[handno] = 0

    res = dds.SolveAllBoards(ctypes.pointer(bo), ctypes.pointer(solved))

    if res != dds.RETURN_NO_FAULT:
        line = ctypes.create_string_buffer(80)
        dds.ErrorMessage(res, line)
        print(f'SolveAllBoards: DDS error: {bo.noOfBoards} {line.value.decode("utf-8")}')
        assert False, grp_start

    for handno in range(0, grp_count):
        r = stdhrs.iloc[grp_start+handno]
        #display(r)
        ref = r['ref'][1]
        dealer = r['Dealer']
        assert dealer in 'NESW'
        vul = r['Vul']
        assert vul in ['None','N_S','E_W','Both'], r
        par = r['Par']
        assert type(par) is tuple, r
        assert len(par) == 2, r
        assert type(par[0]) is int, r
        assert type(par[1]) is list, r
        assert len(par[1]) > 0, r

        #match = True # functions.ComparePlay(ctypes.pointer(solved.solved[handno]), handno)
        functions.PrintPBNHand(f'Group:{n} Hand:{grp_start+handno+1}/{nhands} ACBL Ref:{ref} Vul:{vul}', bo.deals[handno].remainCards)
        line = f"target:{bo.target[handno]} solutions:{bo.solutions[handno]} mode:{bo.mode[handno]}"
        functions.PrintFut(line, ctypes.pointer(solved.solvedBoards[handno]))
```

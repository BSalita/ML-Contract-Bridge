```python
# takes 15s
# read acbl results df
acbl_board_results_file = acblPath.joinpath('acbl-board-results-cleaned.pkl')
acbl_cgdf = pd.read_pickle(acbl_board_results_file)
acbl_cgdf
```

```python
# takes 45s
# todo: strip leading/trailing "" earlier in pipeline.
# todo: is this faster? acbl_playerd = acbl_cgdf.set_index(d+'Num').to_dict()[d+'Name']
# note: striping away leading/trailing "" and (swap names) but can't figure out how to do both in one regex! Spent a lot of time but got nowhere.
acbl_playerd = {num:name for d in mlBridgeLib.NESW for num,name in zip(acbl_cgdf[d+'Num'], acbl_cgdf[d+'Name'].str.replace(r'^"(.*)"$',lambda m: m.group(1),regex=True).str.replace(r'^(.*)\(swap names\)$',lambda m: m.group(1),regex=True))}
len(acbl_cgdf),len(acbl_playerd),acbl_playerd
```

```python
# takes 1s
acbl_player_name_dict = acblPath.joinpath('acbl_player_name_dict.pkl')
with open(acbl_player_name_dict,'wb') as f:
    pickle.dump(acbl_playerd,f)
```

```python
# takes 8m
# create dictionary of results across all clubs. similar to tcg.
# 'Score' must already have non-numerics removed e.g. no AVE,PASS,NP,AVE-,AVE+,[-][0-9]+[FGHL]
acbl_cgd = {}
eg = acbl_cgdf[['EventBoard','EventBoardScore','Score']].groupby('EventBoard')
for eb,ebg in eg:
    ebg_len = len(ebg)
    ebs_l = []
    beats = 0
    assert eb not in acbl_cgd
    for ebs,ebsg in ebg.groupby('Score'):
        assert ebsg['EventBoardScore'].nunique() == 1
        same = len(ebsg)
        mp = beats+(same-1)/2
        pct = mp/(ebg_len-(ebg_len > 1))
        l = [ebsg['Score'].iloc[0],beats,same,mp,pct]
        ebs_l.append(l)
        acbl_cgd[ebsg['EventBoardScore'].iloc[0]] = l
        beats += same
    assert beats == ebg_len
    acbl_cgd[eb] = [ebg_len,ebs_l]
display(len(acbl_cgd))
display(acbl_cgd)
```

```python
acbl_common_game_dict_file = acblPath.joinpath('acblcgd.pkl')
with open(acbl_common_game_dict_file,'wb') as f:
    pickle.dump(acbl_cgd,f)
```
